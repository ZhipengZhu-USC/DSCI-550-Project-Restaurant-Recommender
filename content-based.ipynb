```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import linear_kernel
```

```python
business_dummies = pd.read_csv('business_dummies.csv')
business_full = pd.read_csv('business_full')
```

```python
todrop_cols =[]
for i in business_dummies.columns:
    if ('BestNights' in i) or ('DietaryRestrictions_' in i):
        todrop_cols.append(i)
business_dummies = business_dummies.drop(todrop_cols,axis =1)   
```

```python
business_dummies.sum()
business_dummies.index = business_dummies['business_id']
```

```python
lowfreq_cols =[]
business_dummies_num = business_dummies.drop(['business_id','Unnamed: 0'], axis=1)
for x in business_dummies_num.columns:
    if business_dummies[x].sum()< 200: 
        lowfreq_cols.append(x)

business_dummies_highfreq = business_dummies_num.drop(lowfreq_cols, axis=1)
```

```python
def GetName(id_arr):
    names=[]
    for x in id_arr:
        name = business_full[business_full['business_id'] == x]['name'].values[0]
        names.append(name)
    return names
```

```python
k=5
business_input ='AaIE0Lh51wBrGElQRqhydg'
input_name = GetName([business_input])
input_name = business_full[business_full['business_id'] == business_input]['name'].values[0]
```

```python
item_similarity = pd.DataFrame(cosine_similarity(business_dummies_highfreq), index = business_dummies.business_id,columns = business_dummies.business_id )
top_k_recommend = item_similarity[business_input].sort_values()[:5].index
top_k_name = GetName(top_k_recommend)
print('Your input restaurant: ', input_name)
print()
print('The top {} restuarants recommended to you: \n{}'.format(k, top_k_name))

```

```python
import numpy as np
def evaluate(business, similarity):
    pred = similarity.dot(business)
    pred = pred/ np.array([np.abs(similarity).sum(axis=1)]).T
    return pred
```

```python
item_prediction = evaluate(business_dummies_highfreq,item_similarity )
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
def rmse(prediction, base):
    return sqrt(mean_squared_error(prediction, base))
```

```python
rmse(item_prediction, business_dummies_highfreq)
```
